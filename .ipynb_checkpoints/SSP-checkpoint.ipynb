### SEPSIS SURVIVAL PREDICTION 

## </u>Introduction

Sepsis is a life-threatening medical condition that occurs when the body’s response to infection causes tissue damage and organ failure. Early identification of patients at high risk of death from sepsis is critical to improving treatment outcomes and saving lives. This project aims to develop a machine learning–based predictive model that can estimate the likelihood of survival in sepsis patients using basic demographic and clinical features.

The dataset used in this project, sourced from Kaggle, <u>https://www.kaggle.com/datasets/joebeachcapital/sepsis-survival-minimal-clinical-records</u> consisting of three cohorts; *primary*, *study*, and *validation* — containing patient information such as age, sex, episode number, and hospital outcome (alive or dead). The project follows a complete machine learning pipeline, including data preprocessing, exploratory data analysis (EDA), feature engineering, model training, evaluation, and explainability.

Several models will be trained and compared, including Logistic Regression, Random Forest, and XGBoost, to determine which algorithm provides the most accurate and interpretable predictions. Model evaluation will be based on performance metrics such as Accuracy, Precision, Recall, F1-score, and ROC-AUC. Finally, the best-performing model will be deployed using a simple web interface (Flask or Streamlit), allowing users to input patient details and obtain survival predictions.

By the end of this project, the goal is to demonstrate how AI and machine learning can be used to support clinical decision-making, enhance risk assessment, and potentially contribute to improving patient outcomes in sepsis care.

### Dataset Features and their Meaning

- **age_years:** Describes the patient’s age in years at the time of hospital admission.
Type: Numerical (integer)
Purpose: Older age is typically associated with higher risk and lower survival rates in sepsis patients.

- **sex_0male_1female:** Describes the patient’s biological sex.
Values:
0 = Male
1 = Female
Purpose: To analyze potential gender-based differences in sepsis outcomes.

- **episode_number:** Describes the number of sepsis-related hospital admissions or clinical episodes for the same patient.
Purpose: Repeated episodes may indicate chronic conditions or severe sepsis history, which can influence survival probability.

- **hospital_outcome_1alive_0dead:** Describes the final outcome of the hospital stay for the patient.
Values:
1 = Alive (patient survived)
0 = Dead (patient did not survive)
Purpose: This is the label the model will predict — whether the patient survived sepsis or not.

 #### Feature Engineering 
**Performing some feature engineering to help our model learn new patterns.**

```python
# Feature Engineering: Age Grouping
bins = [0, 18, 40, 60, 80, 120]
labels = ['Child', 'YoungAdult', 'MiddleAge', 'Senior', 'Elderly']

study_df['age_group'] = pd.cut(study_df['age_years'], bins=bins, labels=labels, right=False)
```

```python
# Verify the new feature
print(study_df[['age_years', 'age_group']].head(10))
```

***Observation on Feature Engineering (Age Grouping)***
A new feature, age_group, was successfully created by categorizing patients into five meaningful age ranges:
Child: 0–17 years
YoungAdult: 18–39 years
MiddleAge: 40–59 years
Senior: 60–79 years
Elderly: 80–120 years
From the sample output, patients are correctly assigned to their respective groups (e.g., ages 7 and 17 as Child, 70 and 76 as Senior, and 89 as Elderly).

***Now we will encode our new feature numerically for the model***

```python
# Initialize encoder
le = LabelEncoder()

# Encode the age_group column BEFORE splitting
study_df['age_group_encoded'] = le.fit_transform(study_df['age_group'])

# Display first few rows to confirm encoding
print(study_df[['age_years', 'age_group', 'age_group_encoded']].head(15))

# View what label corresponds to each encoded value
print("\nLabel Mapping:")
for label, code in zip(le.classes_, range(len(le.classes_))):
    print(f"{label} → {code}")
```

***Observation:***
The dataset now contains fully numeric features. Ages range from young (5 years) to elderly (93 years). Gender is encoded as 0 = male and 1 = female. The age_group feature has been successfully transformed from text categories into numeric codes (0–3), making the data ready for model training.

```python
# Define features (X) and target (y)
X = study_df[['age_group_encoded', 'sex_0male_1female', 'episode_number']]
y = study_df['hospital_outcome_1alive_0dead']
```

### Baseline Models Comparison 

```python
# Re-train all models on the imbalanced data 
models = {
    "Logistic Regression (Imbalanced)": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest (Imbalanced)": RandomForestClassifier(random_state=42),
    "XGBoost (Imbalanced)": XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=42),
    "Support Vector Machine (Imbalanced)": SVC(kernel='rbf', probability=True, random_state=42),
    "Decision Tree (Imbalanced)": DecisionTreeClassifier(random_state=42)
}

# Store results
results = []

# Loop through all models
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else [0]*len(y_pred)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_proba)

    results.append({
        "Model": name,
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1,
        "ROC AUC": roc_auc
    })

# Display comparison table
results_df = pd.DataFrame(results).sort_values(by="F1 Score", ascending=False)
print("\n=== Model Performance Summary (Imbalanced Data) ===")
display(results_df)
```

***Observation on baseline trained Model:*** The results show that all five models (Logistic Regression, Random Forest, XGBoost, SVM, and Decision Tree) display similar performance patterns, characterized by:

High Accuracy (≈ 0.81) and High Recall (≈ 1.0) — meaning the models correctly predict nearly all “Alive” patients.
Identical Precision and F1 Scores across models, indicating that the imbalance in the dataset causes the models to heavily favor the majority class (Alive).

ROC AUC values around 0.5–0.57, suggesting poor discriminatory ability — the models cannot effectively distinguish between the two classes (Alive vs. Dead).

The confusion matrices confirm this bias: all models predict almost all samples as the majority class (Alive), leaving zero predictions for the minority class (Dead).

- **In summary:**
Although the models appear to perform well in terms of accuracy, this performance is misleading. The high recall and F1 scores are driven by the dominance of the “Alive” class. Therefore, these results highlight the critical need to handle class imbalance before model training using resampling methods such as SMOTE to ensure fairer and more meaningful model performance

#### Handling class imbalance to improve our Classification Models 
- The dataset is highly imbalanced. Around 81% of patients survived and only 19% died.
This imbalance has cause the model to favor the majority class (survivors) and perform poorly at identifying deaths.
Hence we will perform SMOTE (Synthetic Minority Over-sampling Technique) to balance the minority class then train our Logistic regressor again 

- #### SMOTE (Synthetic Minority Over-sampling Technique)

```python
from imblearn.over_sampling import SMOTE

# Recreate SMOTE-balanced dataset
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print("Before SMOTE:\n", y_train.value_counts())
print("\nAfter SMOTE:\n", y_train_res.value_counts())
print("\nSMOTE balancing completed successfully!")
```

***Observation:*** 
After applying SMOTE, the dataset became perfectly balanced, with 12,355 samples each for the Alive and Dead classes. This balance ensures that the models can learn from both classes equally, reducing bias toward the majority class and improving prediction reliability.

- #### Training the models on the SMOTE Data

### 1. Logistic Regression (SMOTE)

```python
# Initialize model
log_reg_smote = LogisticRegression(random_state=42, max_iter=1000)

# Train model on SMOTE data
log_reg_smote.fit(X_train_smote, y_train_smote)

# Make predictions on test set
y_pred_smote = log_reg_smote.predict(X_test)
```

```python
# Evaluate
print("\n=== Logistic Regression (SMOTE) ===")
print(f"Accuracy: {accuracy_score(y_test, y_pred_smote):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_smote):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_smote):.4f}")
print(f"F1 Score: {f1_score(y_test, y_pred_smote):.4f}")
print(f"ROC AUC: {roc_auc_score(y_test, log_reg_smote.predict_proba(X_test)[:,1]):.4f}")
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_smote)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix - Logistic Regression (SMOTE)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

### 2. Random Forest (SMOTE)

### Logistic Regression Hyperparameter Optimization using GridSearchCV (SMOTE Data)

```python
import warnings
warnings.filterwarnings('ignore')  # Suppress all sklearn warnings

# Define model
log_reg = LogisticRegression(max_iter=500, random_state=42)

# Define hyperparameter grid
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],        
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],  
    'solver': ['liblinear', 'saga'],      
    'l1_ratio': [0, 0.5, 1]              
}

# Initialize GridSearchCV
grid_search = GridSearchCV(
    estimator=log_reg,
    param_grid=param_grid,
    scoring='f1',        
    cv=5,                 
    verbose=1,
    n_jobs=-1
)

# Fit GridSearch on SMOTE-balanced data
grid_search.fit(X_train_res, y_train_res)

# Get best parameters
print("\n=== Best Parameters for Logistic Regression (SMOTE) ===")
print(grid_search.best_params_)

# Evaluate best model on test set
best_log_reg = grid_search.best_estimator_
y_pred_best = best_log_reg.predict(X_test)

# Performance metrics
best_accuracy = accuracy_score(y_test, y_pred_best)
best_precision = precision_score(y_test, y_pred_best)
best_recall = recall_score(y_test, y_pred_best)
best_f1 = f1_score(y_test, y_pred_best)
best_roc_auc = roc_auc_score(y_test, y_pred_best)

print("\n=== Tuned Logistic Regression (SMOTE) ===")
print(f"Accuracy: {best_accuracy:.4f}")
print(f"Precision: {best_precision:.4f}")
print(f"Recall: {best_recall:.4f}")
print(f"F1 Score: {best_f1:.4f}")
print(f"ROC AUC: {best_roc_auc:.4f}")


```
